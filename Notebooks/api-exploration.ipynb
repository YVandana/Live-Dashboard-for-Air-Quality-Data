```python
# 'env : . ..\.venv\Scripts\activate'
import json

from openaq import OpenAQ
```

```python
with open("../secrets.json") as f:
    secrets = json.load(f)
    f.close()
openaq_key = secrets["openaq-api-key"]
```

```python
client = OpenAQ(api_key=openaq_key)
client
```

```python
# Delhi Area 76.486816,28.130128,77.739258,28.878350
# London Area -0.650940,51.280817,0.384521,51.686180
# Brighton Area -0.208740,50.805935,0.005493,50.885709

del_bbox_search_locations = client.locations.list(
    bbox = (76.486816,28.130128,77.739258,28.878350),
    limit = 1000
)

del_bbox_search_locations.meta
```

```python
del_bbox_search_locations.results
```

```python
del_bbox_search_locations.results[0].__dict__
```

```python
locations_info =  {}

for location in del_bbox_search_locations.results:
    locations_info[location.id]= location.name
locations_info
```

```python
len(locations_info)
```

```python
with open("../Locations/del-locations.json", "w", encoding='utf-8') as f:    
    json.dump(locations_info, f, ensure_ascii=False, indent=4)
    f.close()
```
